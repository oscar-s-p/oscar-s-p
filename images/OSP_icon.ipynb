```python
import svgwrite

nx, ny = 300, 300
fontsize = nx/300*120

all_shift = ny*0.05
text_pos = (int(nx*0.5), int(ny*0.63 + all_shift))

r_ellipse = (nx/300*150, nx/300*70)
ellipse_center = (int(nx/2), int(ny*0.48 + all_shift))

dot_pos = (int(nx*0.8), int(ny*0.23 + all_shift))

dwg = svgwrite.Drawing("osp_orbit.svg", 
                       size=(str(nx)+"px", str(ny)+"px"),
                       viewBox=("0 0 "+str(nx)+" "+str(ny)))

# Orbit
dwg.add(dwg.ellipse(
    center=ellipse_center,
    r=r_ellipse,
    fill="none",
    stroke="gray",
    stroke_width=4,
    transform="rotate(-15 "+str(ellipse_center[0])+' '+str(ellipse_center[1])+")"
))

# Planet
dwg.add(dwg.circle(center=dot_pos, r=10, fill="gray"))

# Text
dwg.add(dwg.text(
    "OSP",
    insert=text_pos,
    text_anchor="middle",
    font_size=fontsize,
    font_weight="bold",
    font_family="Arial",
    fill="gray"
))

dwg.save()

import cairosvg

sizes = [32, 180, 192, 512]

for s in sizes:
    cairosvg.svg2png(
        url="osp_orbit.svg",
        write_to=f"osp_orbit-{s}x{s}.png",
        output_width=s,
        output_height=s
    )


from PIL import Image

#sizes = [16, 32, 48, 64, 128, 256]

images = []
for s in sizes:
    img = Image.open(f"osp_orbit-{s}x{s}.png")
    images.append(img)

images[0].save(
    "osp_orbit.ico",
    format="ICO",
    sizes=[(s, s) for s in sizes]
)
```
